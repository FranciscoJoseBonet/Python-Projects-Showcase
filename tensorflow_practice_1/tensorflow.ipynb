```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Churn.csv') # Import the csv as a DataFrame
```

```python
df.head(5) # Here we can see the first five rows of the DataFrame
```

```python
n=pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))             # Convert the categorical variables into boolean data
n = n.applymap(lambda x: 1 if x == True else (0 if x == False else x))  # Convert boolean values to a binary, that can be easily processed by machine learning algorithms
s=df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)                     # We define 's' as target variable
```

```python
n_train, n_test, s_train, s_test=train_test_split(n,s,test_size=0.2)  # Split the total data in two big groups, test and train (for the target and the rest of the data)
```

```python
n_train.head(5)
```

```python
s_train.head(5)
```

```python
from tensorflow.keras.models import Sequential, load_model      # Dependencies
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model=Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(len(n_train.columns),)))      # We are adding hidden layers (Dense) of neurones with different forms of activation function
model.add(Dense(units=64, activation='relu'))                                           # The units are the quantity of neurones for layer         
```

***Final Layer Shape***

- This layer defines how our output will look like. If we only declarate one variable we have one target only
- Sigmoid function give us a binary output

```python
model.add(Dense(units=1, activation='sigmoid'))
```

***Compiling the model***

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

## Fit the model and evaluate the training

```python
model.fit(n_train, s_train, epochs=200, batch_size=32)      # Here we are fitting our model giving it the features velues an the target ones
```

***Prediction***

```python
s_predict=model.predict(n_test)
s_predict=[0 if value <0.5 else 1 for value in s_predict]   # Prediction calculation of the test sample
```

***Accuracy Score***

```python
accuracy_score(s_test, s_predict)   # Test the accuracy to have an evaluation of our method
```

## We can save the model and relod it later

- This code will spawn a new file that has '.keras' format (there also exist '.h5' format but it's considerate deprecated)

```python
model.save('tfmodel.keras')
```
